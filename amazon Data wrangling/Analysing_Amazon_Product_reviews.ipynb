# Sentiment anaylisis
 * sentiment analysis can be determined in 2 ways
    1 By using overall and putting values a positive(4,5), neutral(3) or negative(1,2)
    2 By using review Text and analysing their tone 
I have used the second method it scans the real reviews pf the reviewer

To perform this I will use NLTK (Natural Language Toolkit) platform
and SentimentIntensityAnalyzer

* First I will import them for use

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

Now, I will define function SentimentIntensityAnalyzer as vader

```python
vader = SentimentIntensityAnalyzer()
```

Now I will use polarity scores to assign positive, negative and neutral values to each statement and this will give out how much out of 1, a sentence is positive, neg or nutral which we will use in polarity scores to give out scores.

```python
df['scores'] = df['reviewText'].apply(lambda reviewText: vader.polarity_scores(reviewText))
df['scores']
```

Now we will assign POSITIVE, NEGATIVE and NEUTRAL to gotten score above.

```python
df['compound'] = df['scores'].apply(lambda score_dict: score_dict['compound'])
df['sentiment_type']=''
df.loc[df.compound>0,'sentiment_type']='POSITIVE'
df.loc[df.compound==0,'sentiment_type']='NEUTRAL'
df.loc[df.compound<0,'sentiment_type']='NEGATIVE'
```

Then I will place a Bar Graph using those assigned values

```python
df.sentiment_type.value_counts().plot(kind='bar')
plt.title('Sentiment analysis')
```

# Web Scaraping
### what is web scaraping
- Web scraping is the technique of automatically obtaining structured web data. Web data extraction is another name for it. Pricing monitoring, price intelligence, news monitoring, lead creation, and market research are just a few of the numerous applications of web scraping.

Web scraping is getting data from an external link or html and utilizing it in data format
 - In this I will use the few Products 
 - Then use amazon website to wetch htmls of these products
 - And then find out Its diffrent attributes like Name, Prices, weight and Availibility

For this I will use BeautifulSoup It is a most commonly used web scrapper

```python
from bs4 import BeautifulSoup
import requests
```

The products I selected are as asin

```python
asin = ['B09DKZYBR5','B098QGLXXY','B09BQV32W2']
data = []

for i in asin:
    page = requests.get("https://www.amazon.com.au/dp/" + str(i))
    soup = page.content
    doc = BeautifulSoup(soup, "html.parser")
    
    name = doc.find('span', class_="a-size-large product-title-word-break").string
    price = doc.find('span', class_="a-offscreen").string
    weight = doc.find('td', class_="a-color-secondary a-size-base prodDetSectionEntry", text=' Item Weight ').text[-9:-1]
    availability = doc.find('td', class_="a-color-secondary a-size-base prodDetSectionEntry", text=' Date First Available ').text[-15:-1]
    
    data.append({"asin": i, "name": name, "price": price, "weight": weight, "availability": availability})
    
df = pd.DataFrame(data)
```
